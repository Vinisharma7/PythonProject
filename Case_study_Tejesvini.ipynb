```python
# Items available in the store
items = {
    "Biscuits": (5, 2.50),
    "Cereals": (10, 3.50),
    "Chicken": (20, 13.30),
    "Oats": (20, 3.90),
    "Rice": (30, 4.50)
}

# Delivery charges based on distance (distance: charge)
delivery_charges = {
    5: 10,
    10: 12,
    15: 14
}


# Function to calculate delivery cost based on distance
def calculate_delivery_cost(distance):
    for limit, charge in delivery_charges.items():
        if distance <= limit:
            return charge
    return 20  # Default charge if no distance range applies


# Function to display available items
def display_items(items):
    print("ESSENTIAL ITEMS AVAILABLE ::\n")
    print("%-10s %-8s %-8s %-10s" % ("Item", "Quantity", "Cost", "Item"))
    for item, (quantity, cost) in items.items():
        print("%-10s %-8s %-8s %-10s" % (item, quantity, f"${cost:.2f}", item))
    print("______________________________________________________________")
              
       

#Function to get customer details
def get_customer_details():
    name = input("Fill in your delivery details Enter your name: ")
    distance = float(input("Enter the distance of your delivery address (in miles): "))
    return name, distance

# Function to update purchased items and calculate total cost
def update_items_purchased(items, total_cost):
    purchased_items = {}
    for item, (quantity, cost) in items.items():
        purchased_quantity = int(input(f"How many packets of {item.lower()} are required? "))
        if purchased_quantity > quantity:
            print(f"Sorry, only {quantity} {item} available.")
        else:
            total_cost = (purchased_quantity * cost) + total_cost
            purchased_items[item] = purchased_quantity
            items[item] = (quantity - purchased_quantity, cost)  # Update remaining quantity
    return purchased_items, total_cost


# Function to display the final bill
def display_final_bill(customer_name, purchased_items, total_cost, delivery_cost):
    print("Final Bill:")
    print(f"Customer Name: {customer_name}")
    print("Items Purchased:")
    for item, quantity in purchased_items.items():
        print(f"{item}: {quantity}")
    print(f"Total Cost of Items: ${total_cost:.2f}")
    print(f"Delivery Charges: ${delivery_cost:.2f}")
    print(f"Total Amount Payable: ${total_cost + delivery_cost:.2f}")
```

```python
calculate_delivery_cost(distance)
display_items(items)
get_customer_details()
update_items_purchased(items, total_cost)
display_final_bill(customer_name, purchased_items, total_cost, delivery_cost)
```
